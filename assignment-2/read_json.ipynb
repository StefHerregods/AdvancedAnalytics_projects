```python
import requests

# URL of the JSON file
url = "http://seppe.net/aa/assignment2/dataset.json"

# Download the JSON data from the URL
response = requests.get(url)

# Save the data to a file
with open("dataset.json", "w") as f:
    f.write(response.text)
```

```python
import pandas as pd

# Load the JSON data into a variable (assuming the data is stored in a file called 'data.json')
with open('dataset.json', 'r') as f:
    data = f.read()

# Convert the JSON data into a pandas dataframe
df = pd.read_json(data)

# Print the dataframe to see the contents
print(df)

# Save the dataframe to a CSV file
df.to_csv('dataset.csv', index=False)
```

```python
# Count number of rows per cuisines type
cuisine_list = df['cuisine_labels'].value_counts()
print(cuisine_list)

# plot histogram of number of rows per cuisine type
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.bar(cuisine_list.index, cuisine_list.values)
plt.xticks(rotation=90)
plt.xlabel('Cuisine type')
plt.ylabel('Number of rows')



```

```python
import pandas as pd

# Load the JSON data into a pandas dataframe (assuming the data is stored in a file called 'data.json')
df = pd.read_json('dataset.json')

# Define a function to extract the 'label' value from the cuisines list
def get_cuisine_label(cuisines_list):
    return [cuisine['label'] for cuisine in cuisines_list]

# Apply the function to the 'cuisines' column to create a new column called 'cuisine_labels'
df['cuisine_labels'] = df['cuisines'].apply(get_cuisine_label)

# Print the dataframe to see the new column
print(df)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('..\\assignment-1\\data\\train.csv')
```

```python
import sklearn
import numpy as np
```

```python
df['locatie'] = np.where(df['property_lat'] > 51, 0, 1)
print(df)
```

```python
# Define the threshold for outlier detection
threshold = 3

# Calculate the mean and standard deviation of the data
mean = np.mean(df.property_max_guests)
std = np.std(df.property_max_guests)

# Calculate the Z-score for each data point
z_scores = np.abs((df.property_max_guests - mean) / std)

# Identify the indices of the data points with Z-scores above the threshold
outlier_indices = np.where(z_scores > threshold)

# Remove the outliers from the data
clean_data = np.delete(df.property_max_guests, outlier_indices)

# Print the original and cleaned data sets
print("Original data:", df)
print("Cleaned data:", clean_data)
```
